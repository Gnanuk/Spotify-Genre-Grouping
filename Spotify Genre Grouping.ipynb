```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
# Assuming you have a CSV file named 'spotify_data.csv' containing the necessary data
data = pd.read_csv('spotify dataset.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Dropping unnecessary columns
data = data.drop(['track_id', 'track_name', 'track_artist', 'track_album_id', 'track_album_name', 'track_album_release_date'], axis=1)

# Check for missing values
print("Missing values:\n", data.isnull().sum())

# Remove rows with missing values
data = data.dropna()

# Check the shape of the data
print("Shape of data after preprocessing:", data.shape)
```

```python
# Visualizing the distribution of features
plt.figure(figsize=(10, 6))
sns.histplot(data['loudness'], bins=30, kde=True)
plt.title('Distribution of loudness')
plt.xlabel('loudness')
plt.ylabel('Frequency')
plt.show()

# Visualizing correlation matrix of features
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Features')
plt.show()
```
